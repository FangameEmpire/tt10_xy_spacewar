```python
# Emulation settings
GRID_SIZE = (32, 32) # Y, X

# Points
POINT_A = (15, 16) # Y, X
POINT_B = (15, 23) # Y, X

# Create grid
grid = np.zeros([GRID_SIZE[0], GRID_SIZE[1]])

# Reset grid
def reset_grid(grid):
    grid = np.zeros([GRID_SIZE[0], GRID_SIZE[1]])
    return grid

# Plot grid in photos app
def show_grid(grid):
    # Convert grid to image
    grid_view = Image.fromarray(grid)
    
    # Display image
    grid_view.show()
```

```python
def draw_line(grid, point_a, point_b):
    for y_cur in range(0, GRID_SIZE[0]):
        for x_cur in range(0, GRID_SIZE[1]):
            pass

def bresenham_classic(grid, point_a, point_b):
    # Determine top point
    point_low = (0, 0)
    point_high = (0, 0)
    if (point_a[0] > point_b[0]):
        point_low = point_b
        point_high = point_a
    else:
        point_low = point_a
        point_high = point_b
    print("Bottom point: " + str(point_low) + ", Top point = " + str(point_high))
    
    # Determine deltas
    dx = point_high[1] - point_low[1]
    dy = point_high[0] - point_low[0]
    print("dy = " + str(dy) + ", " + "dx = " + str(dx))
    
    
    # Initial distance calculation
    D_cur = (2 * dy) + dx
    y_cur = point_low[0]
    
    # Store X sweep direction
    x_low_bound = point_low[1]
    x_high_bound = point_high[1]
    x_dir = 0
    
    # Determine X seep direction
    if (x_low_bound < x_high_bound):
        x_dir = 1
    else:
        x_dir = -1
        
    # Determine slope
    y_incr = 1
    
    # Iterate through X
    print("X sweep: " + str(x_low_bound) + " " + str(x_high_bound) + " " + str(x_dir))
    for x_cur in range(x_low_bound, x_high_bound, x_dir):
        print("y_cur = " + str(y_cur) + ", x_cur = " + str(x_cur) + ", D_cur = " + str(D_cur))
        grid [y_cur, x_cur] = 4
        if (D_cur > 0):
            y_cur += y_incr
            if (dx < 0):
                D_cur += 2 * dx
            else:
                D_cur -= 2 * dx
        D_cur += 2 * dy
    
    # Color endpoints
    grid[point_low] = 5
    grid[point_high] = 5
    print("Top: " + str(point_low))
    print("Bot: " + str(point_high))
```
